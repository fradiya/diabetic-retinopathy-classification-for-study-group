```python
from ultralytics import YOLO

model = YOLO('yolo11n-cls.pt')
model.train(  
    data=dataset.location,
    epochs=5,
    imgsz=224,
    batch=64,
    name="ddr-224-nano",
    amp=False,
    device=0,
    )
```
